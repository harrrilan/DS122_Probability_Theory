## **Code**

### Preliminaries

```python
# Import the packages
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  

%matplotlib inline
 
from scipy.stats import expon, norm
```

The PDF for expon is:

$$f(x) = \exp(-x) \text{ for } x \geq 0.$$

This is the “standardized” form of the exponential PDF. It is possible to shift and scale the “standardized” distribution using the loc and scale parameters. A common parameterization for expon is in terms of the rate parameter $\lambda$, such that 
$$f(x) = \lambda \exp(-\lambda x) \text{ for } x \geq 0.$$
This parameterization corresponds to using scale = $\frac{1}{\lambda}$.

```python
# Set the seed for reproducibility
np.random.seed(1)

# Create 100000 random variates from an exponential distribution with rate 0.4
exp_pop = expon.rvs(scale = 1/0.4, size=100000)
```

### Part a

```python
# Visualize the distribution
plt.hist(exp_pop, 400)
plt.title("Exponential Population Distribution")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

### Part b

```python
# Define a function that takes 800 samples of size n from a certain population and returns a list that contains 
# these means
def sampling_distribution(n, population):
    # Setting the seed for reproducibility
    np.random.seed(1)
    # Your code
    # Specify the number of samples
    m = 800
    # Creating an empty list to store sample means
    sample_means = [] 
    # Draw m samples
    for j in range(m):
        # Drawing a sample of size n from the population
        sample = np.random.choice(population, size=n) 
        # Calculating the sample mean
        sample_mean = np.mean(sample) 
        # Appending the sample mean to the list
        sample_means.append(sample_mean)
    return (sample_means)
```

### Part c

```python
# Your code
fig, axs = plt.subplots(2, 2, figsize = (8,12))

sns.histplot(sampling_distribution(5, exp_pop), kde=True, ax = axs[0, 0])
axs[0, 0].set_title('Sample size 5')
sns.histplot(sampling_distribution(25, exp_pop), kde=True, ax = axs[0, 1])
axs[0, 1].set_title('Sample size 25')
sns.histplot(sampling_distribution(50, exp_pop), kde=True, ax = axs[1, 0])
axs[1, 0].set_title('Sample size 50')
sns.histplot(sampling_distribution(100, exp_pop), kde=True, ax = axs[1, 1])
axs[1, 1].set_title('Sample size 100')

for ax in axs.flat:
    ax.set(xlabel='Sample mean')
```

### Part d

For sample size of 50 the approximation of the sampling distribution by the normal distribution becomes acceptable. It has the expected bell shape and mean of 2.5 (=1/0.4).

### Part e

```python
# Your code
# Create 100000 random variates from a normal distribution with mean 500 and standard deviation 75
norm_pop = norm.rvs(500, 75, size=100000)

# Visualize the distribution
plt.hist(norm_pop, 400)
plt.title("Normal Population Distribution")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```

### Part f

```python
# Your code
fig, axs = plt.subplots(2, 2, figsize = (8,12))

sns.histplot(sampling_distribution(5, norm_pop), kde=True, ax = axs[0, 0])
axs[0, 0].set_title('Sample size 5')
sns.histplot(sampling_distribution(25, norm_pop), kde=True, ax = axs[0, 1])
axs[0, 1].set_title('Sample size 25')
sns.histplot(sampling_distribution(50, norm_pop), kde=True, ax = axs[1, 0])
axs[1, 0].set_title('Sample size 50')
sns.histplot(sampling_distribution(100, norm_pop), kde=True, ax = axs[1, 1])
axs[1, 1].set_title('Sample size 100')

for ax in axs.flat:
    ax.set(xlabel='Sample mean')
```

### Part g

For sample size of 5 the approximation of the sampling distribution by the normal distribution becomes acceptable. It has the expected shape and mean of 500.

### Part h

The sample size needed for the exponential population distribution is significantly larger than the one for the normal population distribution. 

The exponential distribution is highly skewed. A large sample size is required for the CLT to approximate the sampling distribution of the sample means by a normal distribution.  

The normal population distribution is symmetric and already has the desired bell shape. Thus, even small samples lead to normal approximation.
